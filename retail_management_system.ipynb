```python
#Making product class for product details
class Product:
    def __init__(self, product_id, name, description, price, stock_quantity):
        self.product_id = product_id
        self.name = name
        self.description = description
        self.price = price
        self.stock_quantity = stock_quantity

    #using accessor method to get details of product id
    def get_product_id(self):
        return self.product_id
    
    #using accessor method to get details of product name
    def get_name(self):
        return self.name
    
    #using accessor method to get details of product description
    def get_description(self):
        return self.description
    
    
    #using accessor method to get details of product price
    def get_price(self):
        return self.price
    
    #using accessor method to get details of product quantity in stock
    def get_stock_quantity(self):
        return self.stock_quantity

    # using mutator methods to modify product name
    def set_name(self, name):
        self.name = name
    
    # using mutator methods to modify product description
    def set_description(self, description):
        self.description = description
    
    # using mutator methods to modify product price
    def set_price(self, price):
        self.price = price

    # using mutator methods to modify product quantity in stock
    def set_stock_quantity(self, stock_quantity):
        self.stock_quantity = stock_quantity

    # using updating method for product details
    def update_product(self, name=None, description=None, price=None, stock_quantity=None):
        if name is not None:
            self.set_name(name)
        if description is not None:
            self.set_description(description)
        if price is not None:
            self.set_price(price)
        if stock_quantity is not None:
            self.set_stock_quantity(stock_quantity)

    # using check method to see whether product is available in stock
    def check_stock(self):
        return self.get_stock_quantity() > 0

```

```python
#Making Customer class for customer details
class Customer:
    def __init__(self, customer_id, name, email):
        self.customer_id = customer_id
        self.name = name
        self.email = email
        self.purchase_history = []

    #using accessor method to get details of customer id
    def get_customer_id(self):
        return self.customer_id

    #using accessor method to get details of custoemr name
    def get_name(self):
        return self.name

    #using accessor method to get details of customer email
    def get_email(self):
        return self.email
    
    #using accessor method to get details of customer purchase history
    def get_purchase_history(self):
        return self.purchase_history

    # using mutators methods to modify customer name
    def set_name(self, name):
        self.name = name
        
    # using mutators methods to modify customer email
    def set_email(self, email):
        self.email = email

    # using update method for customer details
    def update_customer(self, name=None, email=None):
        if name is not None:
            self.set_name(name)
        if email is not None:
            self.set_email(email)

    # using add purchase method to add in customer purchase history
    def add_purchase(self, purchase):
        self.purchase_history.append(purchase)
```

```python
#Making Customer class for customer details
class Sale:
    def __init__(self, sale_id, customer, product, quantity):
        self.sale_id = sale_id
        self.customer = customer
        self.product = product
        self.quantity = quantity
        self.total_price = self.calculate_total_price()

    
    #using accessor method to get details of sale id
    def get_sale_id(self):
        return self.sale_id
    
    #using accessor method to get details of customer id
    def get_customer(self):
        return self.customer

    #using accessor method to get details of product id
    def get_product(self):
        return self.product

    #using accessor method to get details of quantity 
    def get_quantity(self):
        return self.quantity

    #using accessor method to get details of total price
    def get_total_price(self):
        return self.total_price

    #using  mutator method to set the quantity and again calculate the total price
    def set_quantity(self, quantity):
        self.quantity = quantity
        self.total_price = self.calculate_total_price()

    # using calcualte total price method for sale
    def calculate_total_price(self):
        return self.product.get_price() * self.quantity

    # using method to process the sale
    def process_sale(self):
        if self.product.get_stock_quantity() >= self.quantity:
            self.product.set_stock_quantity(self.product.get_stock_quantity() - self.quantity)
            self.customer.add_purchase(self)
        else:
            print("Stock is not enough for sale")
```

```python
# using retail management system method
def retail_management_system():
    products = {}
    customers = {}
    sales = []

    while True:
        print("Following are the options:")
        print("1. Add Product")
        print("2. Register Customer")
        print("3. Process Sale")
        print("4. Generate Reports")
        print("5. Exit")
        choice = input("Choose an option: ")

        if choice == '1':
            # Addition of a new product
            product_id = int(input("Enter product ID: "))
            name = input("Enter product name: ")
            description = input("Enter product description: ")
            price = float(input("Enter product price: "))
            stock_quantity = int(input("Enter stock quantity: "))
            product = Product(product_id, name, description, price, stock_quantity)
            products[product_id] = product
            print(f"Product added successfully: {product_id} - {name}, {description}, ${price}, Stock: {stock_quantity}")

        elif choice == '2':
            # Registeration of a new customer
            customer_id = int(input("Enter customer ID: "))
            name = input("Enter customer name: ")
            email = input("Enter customer email: ")
            customer = Customer(customer_id, name, email)
            customers[customer_id] = customer
            print(f"Customer registered successfully: {customer_id} - {name}, {email}")

        elif choice == '3':
            # Sale processing
            sale_id = int(input("Enter sale ID: "))
            customer_id = int(input("Enter customer ID: "))
            product_id = int(input("Enter product ID: "))
            quantity = int(input("Enter quantity: "))
            customer = customers.get(customer_id)
            product = products.get(product_id)
            if customer and product:
                sale = Sale(sale_id, customer, product, quantity)
                sale.process_sale()
                sales.append(sale)
                print(f"Sale processed successfully: Sale ID: {sale_id}, Customer: {customer.get_name()}, Product: {product.get_name()}, Quantity: {quantity}, Total Price: {sale.get_total_price()}")
            else:
                print("Customer or Product not found.")

        elif choice == '4':
            print("Following are the options")
            print("1. Product Report")
            print("2. Sales Report")
            print("3. Customer Report")
            report_choice = input("Choose an option: ")

            if report_choice == '1':
                # if choice is 1,Generation of a  product report will occur
                print("Product Report:")
                for product in products.values():
                    print(f"ID: {product.get_product_id()}, Name: {product.get_name()}, Stock: {product.get_stock_quantity()}")
            elif report_choice == '2':
                # if choice is 2,Generation of a sales report will occur
                print("Sales Report:")
                for sale in sales:
                    print(f"Sale ID: {sale.get_sale_id()}, Product: {sale.get_product().get_name()}, Quantity: {sale.get_quantity()}, Total: {sale.get_total_price()}")
            elif report_choice == '3':
                # if choice is 3 ,Generation of a customer report will occur
                customer_id = int(input("Enter customer ID for report: "))
                customer = customers.get(customer_id)
                if not customer:
                    print("Customer not found")
                else:
                    print(f"Customer Report for {customer.get_name()}:")
                    for purchase in customer.get_purchase_history():
                        print(f"Sale ID: {purchase.get_sale_id()}, Product: {purchase.get_product().get_name()}, Quantity: {purchase.get_quantity()}, Total: {purchase.get_total_price()}")

        elif choice == '5':
            # finally, if choice is 5 exit the system
            print("Exiting the system.")
            break

        else:
            print("It is not a valid option.")

# calling the function 
retail_management_system()
```
